```python
import pandas as pd
```

```python
chess_games_df = pd.read_csv("chess_games.csv")
chess_games_df
```

```python
opening_moves_summary = []
total_moves = []
white_capture = []
black_capture = []
captures = []
black_response = []
white_response = []
pawn_white_moves = []
pawn_black_moves = []
pawn_moves = []
bishop_white_moves = []
bishop_black_moves = []
bishop_moves = []
knight_white_moves = []
knight_black_moves = []
knight_moves = []
queen_white_moves = []
queen_black_moves = []
queen_moves = []
rook_white_moves = []
rook_black_moves = []
rook_moves = []
king_white_moves = []
king_black_moves = []
king_moves = []
castling_white = []
castling_black = []
castlings = []
check_white = []
check_black = []
checks = []


for i in chess_games_df.itertuples():
    num_opening_moves = 0
    num_white_capture = 0
    num_black_capture = 0
    num_captures = 0
    num_capture_black = 0
    num_capture_white = 0
    num_black_response = 0
    num_white_response = 0
    num_pawn_white_moves = 0
    num_pawn_black_moves = 0
    num_pawn_moves = 0
    num_bishop_white_moves = 0
    num_bishop_black_moves = 0
    num_bishop_moves = 0
    num_knight_white_moves = 0
    num_knight_black_moves = 0
    num_knight_moves = 0
    num_queen_white_moves = 0
    num_queen_black_moves = 0
    num_queen_moves = 0
    num_rook_white_moves = 0
    num_rook_black_moves = 0
    num_rook_moves = 0
    num_king_white_moves = 0
    num_king_black_moves = 0
    num_king_moves = 0
    num_castling_white = 0
    num_castling_black = 0
    num_castlings = 0
    num_check_white = 0
    num_check_black = 0
    num_checks = 0
    moves_summary = i.moves.split()
    total_moves.append(len(moves_summary))
    opening_moves_summary.append(str(moves_summary[0:i.opening_moves]))
    if i.opening_moves % 2 == 1:
        if(len(moves_summary) >= i.opening_moves + 2):
            white_response.append(moves_summary[i.opening_moves])
            black_response.append(moves_summary[i.opening_moves + 1])
        elif(len(moves_summary) >= i.opening_moves + 1):
            white_response.append(moves_summary[i.opening_moves])
            black_response.append(None)
        else:
            white_response.append(None)
            black_response.append(None)
            
    else:
        if(len(moves_summary) >= i.opening_moves + 2):
            white_response.append(moves_summary[i.opening_moves + 1])
            black_response.append(moves_summary[i.opening_moves])
        elif(len(moves_summary) >= i.opening_moves + 1):
            black_response.append(moves_summary[i.opening_moves])
            white_response.append(None)
        else:
            white_response.append(None)
            black_response.append(None)
    for move_num in range(0, len(moves_summary)):
        move = str(moves_summary[move_num])
        if(move.startswith("O")):
            num_castlings += 1
            if(move_num % 2 == 0):
                num_castling_white += 1
            else:
                num_castling_black += 1
        elif(move[0].islower() or move.startswith("P")):
            num_pawn_moves += 1
            if(move_num % 2 == 0):
                num_pawn_white_moves += 1
            else:
                num_pawn_black_moves += 1
        elif move.startswith("K"):
            num_king_moves += 1
            if(move_num % 2 == 0):
                num_king_white_moves += 1
            else:
                num_king_black_moves += 1
        elif move.startswith("B"):
            num_bishop_moves += 1
            if(move_num % 2 == 0):
                num_bishop_white_moves += 1
            else:
                num_bishop_black_moves += 1
        elif move.startswith("Q"):
            num_queen_moves += 1
            if(move_num % 2 == 0):
                num_queen_white_moves += 1
            else:
                num_queen_black_moves += 1   
        elif move.startswith("N"):
            num_knight_moves += 1
            if(move_num % 2 == 0):
                num_knight_white_moves += 1
            else:
                num_knight_black_moves += 1
        elif move.startswith("R"):
            num_rook_moves += 1
            if(move_num % 2 == 0):
                num_rook_white_moves += 1
            else:
                num_rook_black_moves += 1
        if(move.endswith("+")):
            num_checks += 1
            if(move_num % 2 == 0):
                num_check_white += 1
            else:
                num_check_black += 1
        if(move.find("x") != -1):
            num_captures += 1
            if(move_num % 2 == 0):
                num_capture_white += 1
            else:
                num_capture_black += 1
                
    white_capture.append(num_capture_white)
    black_capture.append(num_capture_black)
    captures.append(num_captures)
    pawn_white_moves.append(num_pawn_white_moves)
    pawn_black_moves.append(num_pawn_black_moves)
    pawn_moves.append(num_pawn_moves)
    bishop_white_moves.append(num_bishop_white_moves)
    bishop_black_moves.append(num_bishop_black_moves)
    bishop_moves.append(num_bishop_moves)
    knight_white_moves.append(num_knight_white_moves)
    knight_black_moves.append(num_knight_black_moves)
    knight_moves.append(num_knight_moves)
    queen_white_moves.append(num_queen_white_moves)
    queen_black_moves.append(num_queen_black_moves)
    queen_moves.append(num_queen_moves)
    rook_white_moves.append(num_rook_white_moves)
    rook_black_moves.append(num_rook_black_moves)
    rook_moves.append(num_rook_moves)
    king_white_moves.append(num_king_white_moves)
    king_black_moves.append(num_king_black_moves)
    king_moves.append(num_king_moves)
    castling_white.append(num_castling_white)
    castling_black.append(num_castling_black)
    castlings.append(num_castlings)
    check_white.append(num_check_white)
    check_black.append(num_check_black)
    checks.append(num_checks)
        

print("Your long a$$ for-loop worked!")
```
